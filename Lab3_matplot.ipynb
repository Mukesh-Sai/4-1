# Intro to plotting of graphs

# just need to import matplotlib module to work on graphs

```python
# points plotted on the system and point connected with line graph
import matplotlib.pyplot as plt


x = [1,2,3]
y = [1,4,9]
# this will show only the points scattered onto the graph
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
# this will show only the line onto the graph
plt.plot(x,y)
plt.show()
```

```python
# plotting two graphs onto the same axis
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
#changing the graph line reprsentation
x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x2,y2,"r-.")
plt.show()
```

```python
# plot function x^3
import numpy as np
#x = np.array([1,2,3,4])
x = np.arange(0,5,0.1) # this is arange function where first para is starting_index , then ending_index and last one as incrementer
y = x**3
plt.plot(x,y)
plt.show()
```

```python
a = [3,4,5]   # here in this condition it thinks that array a is of y axis and by default it take the x axis value as [0,1,2,....] up till the values of y is there
plt.plot(a)
plt.show()
```

# customization of graph

```python
import numpy as np

x = np.arange(0,5,0.1) # this is arange function where first para is starting_index , then ending_index and last one as incrementer

y = x**3

plt.plot(x,y, color = "red",marker = "+")

# optional to increase line width  
plt.plot(x,y, color = "red",linewidth = 5,label = "x^3")  #label will tell the name of line as x^3

#here in the plt.plot function the label parameter is used to set the legend i.e., in an order to denote the curve basically what it is and what does it denotes
y2 = x**2

plt.plot(x,y2,color = "blue", linewidth = 5,label = "x^2")

#it is used to add the legend into the axis
plt.legend()   # at the top left the indication of red and blue line with x^3 and x^2 respectively is known as legend

#setting the y label
plt.ylabel("y") 

#setting the x label
plt.xlabel("x")

#setting the title
plt.title("Matplotlib demo")

#construct the axis value according to user defined need

#plt.axis([0,100,0,200])

#ignore
# to create the grid in an order to mark up the values
# and plt.grid() can only work when plt.axis is not in action

plt.grid() # used to set up the grid lines

# add random text
plt.text(2,80,"Text Example", fontsize = 12) # parameters are x and y coordinate and text that need to be placed and text size

#this is used to make the graph visible/show
plt.show()
```

# pie graph

```python
sizes = [3,4,6,2]  #creating a pie graph with the respective sizes

colors = ["blue","red","purple","yellow"] #setting colors in the pies via including the colors in the list

#to add label to each pie in the pie graph
labels = ["A","B","C","D"]

# autopct is the one parameter which show the value accomodated by the pie 

# autopct parameter can have a function which will take the list and convert the value into percentage and return it or can be done via string formatting

# explode is the parameter is the one with which one pie can be highlighted more, so
explode = [0.1,0.2,0,0]  # by altering the value pieces will be scattered

#to show the graph in counter colockwise or not , default value is True
#startangle is the another parameter to set the display of the graph
plt.pie(sizes,colors = colors,labels = labels,autopct = "%.2f", explode = explode, counterclock=False, startangle = 100)

#to add the title
plt.title("Split among classes")

# to make the pie graph in circle shape so - 
plt.axis("equal")

plt.show()
```
