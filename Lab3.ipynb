# data with NAN entries

```python
#creating some nan etries in the data set
import numpy as np
df.iloc[2:4, 1:3] = np.nan  # np.nan does the work to create NaN values
df
```

```python
df.describe()  # overall summary of the dataset
```

```python
#to handle the nan
#by dropping the na data
# use of inplace  = True is there so no need to reassign the dataset
df.dropna(inplace = True)
df.reset_index(drop = True, inplace = True) # reseting the index labels (ie., left most col)
df.head() # looking onto the top 5 rows
```

```python
#instead of dropping na entries changing with other data
#by putting the values onto it

# used in data cleaning

df.iloc[2:4, 1:3] = np.nan  # setting the value as nan from index 2 to 3 and column 1 to 2
print(df) # viewing the dataset
df.sw.fillna(df.sw.mean(), inplace = True)  # replacing the value of nan with the mean value of sw column with inplace = True
print("-----------------------------------------------------------------")
print(df)  # viewing the dataset
# the nan values in sw are filled
```

```python
# another way to fill nan values
df.pl.fillna(df.pl.mean(), inplace = True)
print(df)
```

```python
# get some customized sort of data with the data summary
a = df[df.sl > 4.0]
a.describe()
```

```python
df = x.copy() # copying the original dataset
df.columns = ['sl','sw','pl','pw','flower_type'] # and setting up the columns of dataset
df
```

```python
df = x.copy()  # again copying the original dataset
df.columns = ['sl','sw','pl','pw','flower_type'] # managing the cols in dataset
df  # viewing the data
df["gender"] = "female"  # creating the column as gender with female values in it
df.iloc[0:10, 5] = "male"  # changing the value of gender column from female --> male from index 0 to 9 of the 5the column
df.head(20) # looking onto the top 20 values of dataset
```

```python
#manipulating string based data as converting the string data to integer data as it is easier to manipulate
# apply is an inbuilt function inside pandas
def f(s):
    if(s == "male"):
        return 0
    else:
        return 1

df["sex"] = df.gender.apply(f)
del df["gender"]
df
```
